f) For the matrix: n_array = np.array([[55, 25, 15],[30, 44, 2],[11, 45, 77]]), find
i. Sum of diagonal elements
ii. Eigen values of matrix
iii. Eigen vectors of matrix
iv. Inverse of matrix
v. Determinant of matrix


```python
import numpy as np

n_array = np.array([[55, 25, 15], [30, 44, 2], [11, 45, 77]])

# Sum of diagonal elements
diagonal_sum = np.trace(n_array)

# Eigenvalues and Eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(n_array)

# Inverse of matrix
inverse_matrix = np.linalg.inv(n_array)

# Determinant of matrix
determinant = np.linalg.det(n_array)

print(f"Sum of diagonal elements: {diagonal_sum}")
print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors:\n{eigenvectors}")
print(f"Inverse matrix:\n{inverse_matrix}")
print(f"Determinant of matrix: {determinant}")
```

(g) Multiply the following matrices and also find covariance between matrices using NumPy:
i. p = [[1, 2], [2, 3]]
q = [[4, 5], [6, 7]]
ii. p = [[1, 2], [2, 3], [4, 5]]
q = [[4, 5, 1], [6, 7, 2]]


```python
#i
import numpy as np

p = np.array([[1, 2], [2, 3]])
q = np.array([[4, 5], [6, 7]])

# Matrix multiplication
result_matrix = np.dot(p, q)

# Calculate covariance between matrices
covariance_matrix = np.cov(p, q)

print("Matrix multiplication result:")
print(result_matrix)

print("\nCovariance matrix between matrices:")
print(covariance_matrix)
```

```python
#ii
import numpy as np

p = np.array([[1, 2], [2, 3], [4, 5]])
q = np.array([[4, 5, 1], [6, 7, 2]])

# Matrix multiplication
result_matrix = np.dot(p, q)

# Calculate covariance between matrices
covariance_matrix = np.cov(p, q)

print("Matrix multiplication result:")
print(result_matrix)

print("\nCovariance matrix between matrices:")
print(covariance_matrix)
```

(h) For the matrices: x = np.array([[2, 3, 4], [3, 2, 9]]); y = np.array([[1, 5, 0], [5, 10, 3]]),
find inner, outer and cartesian product?


```python
import numpy as np

x = np.array([[2, 3, 4], [3, 2, 9]])
y = np.array([[1, 5, 0], [5, 10, 3]])

# Inner product (dot product)
inner_product = np.inner(x, y)

# Outer product
outer_product = np.outer(x, y)

# Cartesian product
cartesian_product = np.cartesian([x, y])

print("Inner product (dot product):")
print(inner_product)

print("\nOuter product:")
print(outer_product)

print("\nCartesian product:")
print(cartesian_product)
```

##Q2: Based on NumPy Mathematics and Statistics
(a) For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])
i. Find element-wise absolute value
ii. Find the 25th, 50th, and 75th percentile of flattened array, for each column, for each
row.
iii. Mean, Median and Standard Deviation of flattened array, of each column, and
each row

```python

array = np.array([[1, -2, 3], [-4, 5, -6]])

# i. Element-wise absolute value
abs_array = np.abs(array)
print("Element-wise absolute value:\n", abs_array)

# ii. Percentiles for each column and each row
percentiles = [25, 50, 75]
for axis in [0, 1]:
    print(f"Percentiles for axis {axis} ({'Columns' if axis == 0 else 'Rows'}):")
    for p in percentiles:
        percentile_value = np.percentile(array, p, axis=axis)
        print(f"{p}th percentile:", percentile_value)

# iii. Mean, Median, and Standard Deviation
flattened_array = array.flatten()
mean = np.mean(flattened_array)
median = np.median(flattened_array)
std_dev = np.std(flattened_array)

print("Flattened Array:")
print(flattened_array)

print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std_dev)

# Mean, Median, and Standard Deviation for each column and each row
for axis in [0, 1]:
    print(f"Statistics for axis {axis} ({'Columns' if axis == 0 else 'Rows'}):")
    axis_values = np.take(array, axis=axis, indices=np.arange(array.shape[axis]))
    axis_mean = np.mean(axis_values)
    axis_median = np.median(axis_values)
    axis_std_dev = np.std(axis_values)
    print("Mean:", axis_mean)
    print("Median:", axis_median)
    print("Standard Deviation:", axis_std_dev)
```

i. Sorted array

```python
import numpy as np

# Given array
array = np.array([10, 52, 62, 16, 16, 54, 453])

# i. Sorted array
sorted_array = np.sort(array)
print("i. Sorted array:", sorted_array)

# ii. Indices of sorted array
sorted_indices = np.argsort(array)
print("ii. Indices of sorted array:", sorted_indices)

# iii. 4 smallest elements
smallest_elements = sorted_array[:4]
print("iii. 4 smallest elements:", smallest_elements)

# iv. 5 largest elements
largest_elements = sorted_array[-5:]
print("iv. 5 largest elements:", largest_elements)
```

```python
from PIL import Image
import numpy as np
import os

def img_to_array(path):
    image = Image.open(path)
    image_array = np.array(image)

    if len(image_array.shape) == 3:  # RGB image
        mode = "RGB"
        data = image_array.reshape(-1, 3)  # Flattening the RGB channels
    elif len(image_array.shape) == 2:  # Grayscale image
        mode = "L"
        data = image_array.reshape(-1, 1)  # Flattening the grayscale channel
    else:
        raise ValueError("Unsupported image format")

    output_path = os.path.splitext(path)[0] + "_array.txt"

    with open(output_path, 'w') as f:
        f.write(f"Image Mode: {mode}\n")
        for pixel in data:
            f.write(' '.join(map(str, pixel)) + '\n')

    return output_path

# Example usage
image_path = '/content/pexels-burak-the-weekender-1250664.jpg'
saved_array_path = img_to_array(image_path)
print(f"Image array saved at: {saved_array_path}")
```
